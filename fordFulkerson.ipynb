```python
from collections import deque

infinity = float("inf")


def make_graph():
    
    return [
        [0,20,30,10,0],
        [0,0,40,0,30],
        [0,0,0,10,20],
        [0,0,5,0,20],
        [0,0,0,0,0],

    ]



def bfs(G, bit, bas, parent):
    visited = [False] * len(G)

    queue = deque()
    queue.append(bit)



    visited[bit] = True

    while queue:
        node = queue.popleft()

        for i in range(len(G[node])):
            if visited[i] is False and G[node][i] > 0:
                queue.append(i)
                visited[i] = True
                parent[i] = node





    return True if visited[bas] else False


def ford_fulkerson(G, bit, bas):
    
    parent = [-1] * (len(G))
    max_flow = 0


    while bfs(G, bit, bas, parent):
        path_flow = infinity

        s = bas

        while s != bit:
            
            path_flow = min(path_flow, G[parent[s]][s])
            s = parent[s]

        max_flow += path_flow
        y = bas

        
        while y != bit:
            x = parent[y]
            G[x][y] -= path_flow
            G[y][x] += path_flow
            y = parent[y]


    return max_flow


def main():
    G = make_graph()
    bit = 0
    bas = 4
    max_flow = ford_fulkerson(G, bit, bas)
    print(f'Maximum akış: {max_flow}')
    for i in range(0,5):
        print("{0}. Dugum akis miktarlari:{1}".format(i,G[i]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(1,2,G[1][0]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(1,3,G[2][0]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(1, 4, G[1][4]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(2, 3, G[2][3]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(2, 5, G[4][1]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(3, 4, G[3][0]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(3, 5, G[4][2]))
    print("Baglanti {0}-{1} akis miktari :{2}:".format(4, 5, G[4][3]))




main()
```
